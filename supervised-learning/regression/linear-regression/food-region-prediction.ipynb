```python
confidence = LabelEncoder()
region = LabelEncoder()
```

```python
data['Confidence in estimate'] = confidence.fit_transform(data['Confidence in estimate'])
data['Region'] = region.fit_transform(data['Region'])
```

```python
data.head()
```

```python
features = ['Household estimate (kg/capita/year)', 'Retail estimate (kg/capita/year)', 
            'Food service estimate (kg/capita/year)', 'Confidence in estimate', 'Region']
target = 'combined figures (kg/capita/year)'
```

```python
X = data[features]
y = data[target]
```

```python
X
```

```python
y
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=12)
```

```python
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
model = RandomForestRegressor(random_state=42)
```

```python
model.fit(X_train, y_train)

# Predictions and Evaluation
y_pred = model.predict(X_test)
mse = mean_squared_error(y_test, y_pred)
rmse = mse ** 0.5

# Display the RMSE
print(f'Root Mean Squared Error (RMSE): {rmse}')
```
